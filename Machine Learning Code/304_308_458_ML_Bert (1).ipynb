```python
import pandas as pd
import re
import emoji
def cleanTweets(text):
    text = re.sub(r'@\w+', '', text)                                # Menghapus Username
    text = re.sub('https?:\/\/\S+', '', text)                       # Menghapus Link
    text = re.sub('[^a-zA-Z]', ' ', str(text).lower().strip())      # Lower Case setiap kata
    text = re.sub('#','',text)                                      # Hanya menghapus simbol hastag
    text = re.sub('[!"#$%&\()*+,-./:;<=>?@[\\]^_`{|}~]+','', text)  # Menghapus simbol-simbol beserta string selanjutnya
    text = re.sub('RT[\s]+','',text)                                # Menghapus kata yang memiliki "retweet / RT"
    text = re.sub('\n',' ',text)                                    # Menghapus newline
    text = emoji.replace_emoji(text, replace='')                    # Menghapus emoji atau emoticon
    return text
text['full_text'] = text['full_text'].apply(cleanTweets)
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification, AutoConfig
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
from transformers import create_optimizer
from tensorflow.keras.losses import SparseCategoricalCrossentropy

# --- Persiapan Dataset ---
texts = text['stopword'].values
labels = text['Sentimen'].map({"Others": 0, "Stress": 1, "Tidak Stress": 2}).values

train_texts, temp_texts, train_labels, temp_labels = train_test_split(
    texts, labels, test_size=0.2, random_state=42, stratify=labels
)
val_texts, test_texts, val_labels, test_labels = train_test_split(
    temp_texts, temp_labels, test_size=0.5, random_state=42, stratify=temp_labels
)

print("Distribusi Label:")
print("Train:", dict(pd.Series(train_labels).value_counts()))
print("Validation:", dict(pd.Series(val_labels).value_counts()))
print("Test:", dict(pd.Series(test_labels).value_counts()))

# --- Tokenizer ---
tokenizer = AutoTokenizer.from_pretrained('indolem/indobertweet-base-uncased')

# Tokenisasi dengan panjang dinamis
max_length = 128  # Anda bisa menyesuaikan panjang ini jika perlu
# Convert the lists of tokens back to strings
train_texts_str = [' '.join(text) for text in train_texts]
val_texts_str = [' '.join(text) for text in val_texts]
test_texts_str = [' '.join(text) for text in test_texts]

# Now use these strings for tokenization
train_encodings = tokenizer(train_texts_str, truncation=True, padding=True, max_length=max_length)
val_encodings = tokenizer(val_texts_str, truncation=True, padding=True, max_length=max_length)
test_encodings = tokenizer(test_texts_str, truncation=True, padding=True, max_length=max_length)

# --- Dataset TensorFlow ---
def create_tf_dataset(encodings, labels):
    return tf.data.Dataset.from_tensor_slices((
        {key: tf.constant(val) for key, val in encodings.items()},
        tf.constant(labels)
    ))

train_dataset = create_tf_dataset(train_encodings, train_labels).shuffle(100).batch(8)
val_dataset = create_tf_dataset(val_encodings, val_labels).batch(8)
test_dataset = create_tf_dataset(test_encodings, test_labels).batch(8)

# --- Model Configuration ---
config = AutoConfig.from_pretrained('indolem/indobertweet-base-uncased', num_labels=3)
model = TFAutoModelForSequenceClassification.from_pretrained(
    'indolem/indobertweet-base-uncased',
    config=config,
    from_pt=True
)

# --- Optimizer dan Kompilasi ---
batch_size = 8
num_train_epochs = 5
learning_rate = 5e-5
num_train_steps = len(train_dataset) * num_train_epochs

optimizer, lr_schedule = create_optimizer(
    init_lr=learning_rate,
    num_train_steps=num_train_steps,
    num_warmup_steps=0,
)
model.compile(optimizer=optimizer, loss=SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])

# --- Training ---
history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=num_train_epochs,
    batch_size=batch_size
)

# --- Evaluasi ---
loss, accuracy = model.evaluate(test_dataset)
print(f"Test Loss: {loss}, Test Accuracy: {accuracy}")

# --- Classification Report ---
predicted_logits = model.predict(test_dataset)
predicted_classes = tf.math.argmax(predicted_logits.logits, axis=1).numpy()
print(classification_report(test_labels, predicted_classes, target_names=["Others", "Stress", "Tidak Stress"]))

# --- Visualisasi ---
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.show()

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.show()
```

```python
import pandas as pd

# Decode predicted classes dan label asli ke kategori
true_sentiments = pd.Series(test_labels).map({0: "Stress", 1: "Tidak Stress", 2: "Others"})
predicted_sentiments = pd.Series(predicted_classes[:len(test_labels)]).map({0: "Stress", 1: "Tidak Stress", 2: "Others"})

# Membuat dataset baru
dataset_new = pd.DataFrame({
    "clean_text": test_texts,  # Pastikan 'test_texts' adalah teks asli yang sudah bersih
    "true_sentiment": true_sentiments,
    "predicted_sentiment": predicted_sentiments
})

dataset_new = dataset_new[dataset_new['true_sentiment'] != dataset_new['predicted_sentiment']]

# Menampilkan beberapa baris dari dataset baru
print(dataset_new.head())

# Simpan ke file CSV jika diperlukan
dataset_new.to_csv("Error Analysis indobertweet.csv", index=False)
print("Dataset baru berhasil disimpan ke 'Error Analysis indobertweet.csv'")
```
